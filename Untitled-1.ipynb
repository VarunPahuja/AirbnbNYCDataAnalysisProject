**Data Cleaning**

```python
data.isnull().sum() #sum of total na values 

data.dropna(inplace = True) #dropping all missing values rows |
```

```python
data.shape
```

```python
#dealing with duplicates 
data.duplicated().sum()

data[data.duplicated()]
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data.dtypes
data['id']= data['id'].astype(object)


```

```python
data['hos5_id']= data['host_id'].astype(object)
```

**Data Analysis** 

```python
sns.boxplot(data=data,x='price')
```

```python
#identifying outliers in price
df = data[data['price']<1500]
sns.boxplot(data=df,x='price')
```

```python
#Price Distribution
 
plt.figure(figsize=(8,5))
sns.histplot(data = df,x='price')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
df.dtypes
```

```python
#Availivility 365
 
plt.figure(figsize=(6,3))
sns.histplot(data = df,x='availability_365')
plt.title("Availivility 365")
plt.xlabel("Availivility 365")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
#Feature Engineering (Caldculatiing Price Per Bed)
df['PricePerBed']= df['price']/df['beds']
df.head()
```

```python
df.groupby(by='neighbourhood_group')['PricePerBed'].mean()
```
